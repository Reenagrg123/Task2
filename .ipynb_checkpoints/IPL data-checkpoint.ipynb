```python
# Q.1 Display total runs made by each team during the all seasons till now.
# Arrange this data in the ascending order of their total runs.

team_wise_runs=deliveries_data.groupby('batting_team')['total_runs'].sum().sort_values().reset_index(name="Total runs")
print("Overall team score of all seasons during IPL:")
team_wise_runs
```

```python
# Q.2 List all the matches of 'V Kohli' in which he run out by the bowler 'Z khan'.

V_kohli_matches=deliveries_data.query("batsman == 'V Kohli' and dismissal_kind=='run out' and bowler=='Z Khan'" ).reset_index()
V_kohli_matches

```

```python
# Q.3 Display season wise winners with all the necessary details like both the playing teams and by how many runs 
# winning team run this match?

winners_data=matches_data.groupby('season')['season','team1','team2','winner','win_by_runs'].tail(1).reset_index().sort_values('season')
winners_data
```

```python
# Q.4 capitalize all the names of teams who has won the toss in all matches of all seasons and then display the table in the sorted order of seasons.

capitalize = lambda x: x.upper()
to_be_capitalized=matches_data

to_be_capitalized['toss_winner']=to_be_capitalized['toss_winner'].apply(capitalize)
to_be_capitalized.sort_values('season').reset_index()



```

```python
# Q.5 List out the details of all winning matches of 'Rajasthan Royals'. 

matches_data.query("(team1 == 'Rajasthan Royals' or team2 == 'Rajasthan Royals') and winner=='Rajasthan Royals'" ).reset_index()
```
