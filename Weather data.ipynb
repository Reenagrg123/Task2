```python
# data cleaning

#step-1  Assigning NaN
import pandas as pd

df=pd.read_csv("testset.csv",parse_dates=["datetime_utc"],na_values=["not available","n.a",-9999.0])

# type(df.datetime_utc[0])
df.set_index("datetime_utc",inplace=True)
df

```

```python
# step-2 Adding missing rows


dt=pd.date_range("1996-11-01 11:00:00","2017-04-24 18:00:00",freq="60min")
indx=pd.DatetimeIndex(dt)
df=df.reindex(indx)
df

```

```python
#step-3  interpolating the new added rows
df.fillna(0,inplace=True)
df.fillna(method="ffill",limit=2,inplace=True)
new_df=df.interpolate(method='time')
new_df
 
```
