```python
# data cleaning

#step-1  Assigning NaN
import pandas as pd

df=pd.read_csv("testset.csv",parse_dates=["datetime_utc"],na_values=["not available","n.a",-9999.0])

# type(df.datetime_utc[0])
df.set_index("datetime_utc",inplace=True)
df

```

```python
# step-2 Adding missing rows


dt=pd.date_range("1996-11-01 11:00:00","2017-04-24 18:00:00",freq="60min")
indx=pd.DatetimeIndex(dt)
df=df.reindex(indx)
df

```

```python
#step-3  interpolating the new added rows and detecting and removing outlier

df.fillna(0,inplace=True)
df.fillna(method="ffill",limit=2,inplace=True)
new_df=df.interpolate(method='time')
new_df

 
```

```python
graph_data=new_df.groupby(new_df.index.year).mean()
graph_data
```

```python
import matplotlib.pyplot as plt

def create_plot(x,y):
    if y==' _dewptm':
        y=graph_data[' _dewptm'].tolist()
    elif y==' _fog': 
        y=graph_data[' _fog'].tolist()
        
    elif y==' _hail': 
        y=graph_data[' _hail'].tolist()
    elif y==' _heatindexm': 
        y=graph_data[' _heatindexm'].tolist()    
    elif y==' _rain': 
        y=graph_data[' _rain'].tolist()
    elif y==' _snow': 
        y=graph_data[' _snow'].tolist()
    elif y==' _tempm': 
        y=graph_data[' _tempm'].tolist()
    elif y==' _thunder': 
        y=graph_data[' _thunder'].tolist() 
    elif y==' _tornado': 
        y=graph_data[' _tornado'].tolist()     
    elif y==' _vism': 
        y=graph_data[' _vism'].tolist() 
    elif y==' _wdird': 
        y=graph_data[' _wdird'].tolist()
    elif y==' _wgustm': 
        y=graph_data[' _wgustm'].tolist() 
    elif y==' _windchilm': 
        y=graph_data[' _windchillm'].tolist() 
    elif y==' _wspdm': 
        y=graph_data[' _wspdm'].tolist()     
            
        
    plt.plot(x,y)
    plt.xlabel("Time(in years)")
  
    plt.show()


x=new_df.index.year.unique().tolist()



```
